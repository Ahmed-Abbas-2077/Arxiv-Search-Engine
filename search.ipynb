# Preprocessing

- Tokenization: Split documents into individual words or tokens.
- Lowercasing: Convert all text to lowercase for case insensitivity.
- Stopword Removal: Eliminate common words (e.g., "and", "the", "is") that do not contribute much to the meaning of the document.
- Stemming or Lemmatization: Reduce words to their base or root form (e.g. "running" to "run").

```python
import pandas as pd
dfCsv = pd.read_csv('arxiv-metadata-oai-snapshot.csv', nrows=10000)
dfCsv
```

## Tokenization

```python
import nltk
nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize
```

```python
dfCsv['abstract_token'] = dfCsv['abstract'].apply(word_tokenize)
dfCsv['abstract_token']
```

## Lowercasing

```python
dfCsv['abstract_token'] = dfCsv['abstract_token'].apply(lambda x: [word.lower() for word in x])
dfCsv['abstract_token']
```

## Stopword Removal

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
dfCsv['abstract_token'] = dfCsv['abstract_token'].apply(lambda x: [word for word in x if word not in stop_words])
dfCsv['abstract_token']
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
wordnet = nltk.corpus.wordnet
```

```python
def get_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
wnl = WordNetLemmatizer()
```

```python
def lem(tokens):
    # Tag tokens with POS
    tagged = nltk.pos_tag(tokens)
    
    lemmatized = [
        wnl.lemmatize(word, get_pos(tag))
        for word, tag in tagged
    ]
    
    return lemmatized
```

```python
words = ['running', 'ran', 'runs', 'run']
lem(words)
```

```python
dfCsv['abstract_token'] = dfCsv['abstract_token'].apply(lem)
dfCsv['abstract_token']
```

# Indexing and Retrieval

```python
# turn the list of tokens back into a string
dfCsv['processed_abstract'] = dfCsv['abstract_token'].apply(lambda x: ' '.join(x))
dfCsv['processed_abstract']
```

```python
# creating a new column that is the string version of id
dfCsv['docno'] = dfCsv['id'].astype(str)
dfCsv['docno']
```

```python
docs = []
for i, row in dfCsv.iterrows():
    docs.append({
        "docno": row['docno'],
        "text": row['processed_abstract']
    })
```

```python
def preprocess(query):
    query = word_tokenize(query) # tokenize
    query = [word.lower() for word in query] # lowercase
    query = [word for word in query if word not in stop_words] # remove stopwords
    query = lem(query) # lemmatize
    query = ' '.join(query) # back to string
    return query
```

```python
def search_term(term):
    term = preprocess(term)
    print(f"\n Searching for: {term}")

    try:
        pointer = index.getLexicon()[term]
        for posting in index.getInvertedIndex().getPostings(pointer):
            print(f"Document ID: {posting.getId()}, Length: {posting.getDocumentLength()}")
    except KeyError:
        print(f" Term '{term}' not found in the index.")
```
