# Preprocessing

- Tokenization: Split documents into individual words or tokens.
- Lowercasing: Convert all text to lowercase for case insensitivity.
- Stopword Removal: Eliminate common words (e.g., "and", "the", "is") that do not contribute much to the meaning of the document.
- Stemming or Lemmatization: Reduce words to their base or root form (e.g. "running" to "run").

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
wordnet = nltk.corpus.wordnet
```

```python
def get_pos(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return wordnet.NOUN
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
wnl = WordNetLemmatizer()
```

```python
nltk.pos_tag(["with"])
```

```python
def lem(tokens):
    # tagging each token with their part of speecjh
    tagged = nltk.pos_tag(tokens) # list of tuples (word, tag)
    
    lemmatized = [
        wnl.lemmatize(word, get_pos(tag))
        for word, tag in tagged
    ]
    
    return lemmatized
```

```python
def preprocess(query):
    query = word_tokenize(query) # tokenize
    query = [word.lower() for word in query] # lowercase
    query = [word for word in query if word not in stop_words] # remove stopwords
    query = lem(query) # lemmatize
    query = ' '.join(query) # back to string
    return query
```

```python
def search_term(term):
    term = preprocess(term)
    print(f"\n search term: {term}")

    try:
        pointer = index.getLexicon()[term]
        for posting in index.getInvertedIndex().getPostings(pointer):
            print(f"doc id: {posting.getId()}, len: {posting.getDocumentLength()}")
    except KeyError:
        print(f"'{term}' not found in the index.")
```

# Query Expansion

```python
def PRF(query, top_docs, expansion_number):
    
    query = preprocess(query)
    print(f"\nsearching: {query}")
    tf_idf = pt.BatchRetrieve(index, wmodel="TF_IDF")
    results = tf_idf.search(query)
    
    if len(results) == 0:
        print("no results found.")
        return
    
    rm3_expand = pt.rewrite.RM3(index, fb_docs=top_docs, fb_terms=expansion_number)
    prf_pipeline =  tf_idf >> rm3_expand >> tf_idf # we can skip the two steps below, just felt it more intuitive
    # expanded = prf_pipeline.search(query).iloc[0]["query"].split()[1:]
    # joined = " ".join(expanded)
    final_results = prf_pipeline.search(query)
    
    # print(f"\nwithout PRF, retrieved {len(results)}:\n")
    # for _, row in results.iterrows():
    #     print(f"document id: {row['docno']}, score: {row['score']}")    
    # print(f"\nwith PRF, retrieved {len(final_results)}:\n")
    # for _, row in final_results.iterrows():
    #     print(f"document id: {row['docno']}, score: {row['score']}")
        
    return final_results
```

## Applying Word2Vec for Query Expansion

```python
dfCsv['abstract_tokens'] = dfCsv['abstract'].apply(word_tokenize)
dfCsv['abstract_tokens'] 
```

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=dfCsv["abstract_tokens"], vector_size=100, window=5, min_count=5, sg=1, workers=4, epochs=30)
word_embeddings = model.wv
```

```python
sim = word_embeddings.most_similar("python", topn=10)
print(sim)
```

```python
def expand_query(query, embeddings, topn=3):
    expanded_terms = []
    query = preprocess(query)
    for term in query.split():
        if term in embeddings:
            sim = embeddings.most_similar(term, topn=topn)
            sim_terms = [x[0] for x in sim]
            expanded_terms.extend(sim_terms)
    query += " " + " ".join(expanded_terms)            
    return query
```

```python
expand_query("Quantum Computing Hackathon", word_embeddings, 5)
```

```python
def query_expansion(query, embeddings, topn=100):
    expanded = expand_query(query, embeddings, topn)
    results = PRF(expanded, 100, 10)
    return results
```

```python
query_expansion("Topological Deep Learning", word_embeddings, 5)
```

# Evaluation

```python
queries = [
    "Quantum Computing Hackathon",
    "Topological Deep Learning",
    "Machine Learning",
    "Deep Learning",
    "Natural Language Processing",
    "Reinforcement Learning",
    "Generative Adversarial Networks",
    "Graph Neural Networks",
    "Explainable AI",
    "Federated Learning"
]
```

```python
from tqdm import tqdm
import numpy as np
```

```python
def qrels(queries, docs):
    qrels = []
    
    for id, query in enumerate(queries):
        Qterms = preprocess(query).split()
        
        for i, row in tqdm(docs.iterrows(), total=len(docs)):
            doc_id = row["docno"]
            title = row["title"]
            abstract = row["abstract"]
            text = f"{title}  {abstract}"
            text = preprocess(text).split()
            
            text = set(text)
            Qterms = set(Qterms)
            overlap = len(text.intersection(Qterms))
            
            relevance = 0
            if overlap >= 3:
                relevance = 3
            elif overlap >= 2:
                relevance = 2
            elif overlap >= 1:
                relevance = 1
                
            qrels.append((id, doc_id, relevance))

    return pd.DataFrame(qrels, columns=["query_id", "doc_id", "relevance"])       
```

```python
qrels_df = qrels(queries, dfCsv)
qrels_df 
```

```python
qrels_df.to_csv("qrels.csv", index=False)
qrels_df = pd.read_csv("qrels.csv")
```

```python
query0_relevancy = qrels_df[qrels_df["query_id"] == 0]
query0_relevancy
```

```python
import math
def nDCG(ranked_results, n=10):
    scores = ranked_results["relevance"].iloc[:n]
    
    sum = scores.iloc[0]
    for index, score in enumerate(scores):
        if index == 0:
            continue
        score = score / math.log(index+1, 2)
        sum += score

    sorted_scores = scores.sort_values(ascending=False)  
    optimal_sum = sorted_scores.iloc[0]
    for index, score in enumerate(sorted_scores):
        if index == 0:
            continue
        score = score / math.log(index+1, 2)
        optimal_sum += score
   
    return sum / optimal_sum
```

```python
nDCG(query0_relevancy, 100)
```
